````python
# Determinar si un estudiante se encuentra en nivel Inicio, Proceso o Logrado.
# Inicio < 11
# En Proceso >=11 && <=14
# Logrado >=15

nota1 = 15
nota2 = 11
nota3 = 13

promedio = (nota1 + nota2 + nota3) / 3

if promedio >= 15:
  print("Logrado")
elif promedio >= 11:
    print("En proceso")
else:
    print("Inicio")
````

El dscto se determina de acuerdo al medio de pago:



*   Efectivo -> 10% dscto
*   Tarjeta de débito -> 8% dscto
*   Tarjeta de crédito -> 5% dscto



````python
pago = 100
medioPago = input("Ingresa el medio de pago: ")

if medioPago == "EFECTIVO":
  print(f"Tienes un descuento de {pago*0.1}")
elif medioPago == "TARJETA DE CREDITO":
  print(f"Tienes un descuento de {pago*0.08}")
else:
  print(f"Tienes un descuento de {pago*0.05}")
````

````python
productos = {
    "nombre": ["cuaderno", "libro"],
    "precio": [9, 50],
    "cantidad": [10, 15]
}

nombre = input("Ingresa el producto: ")

if nombre in productos["nombre"]:
  cant_venta = int(input("Ingresa la cantidad vendida: ")) # 3
  indice = productos["nombre"].index(nombre) # nombre.index("cuaderno") = 0
  calculo = productos["cantidad"][indice] - cant_venta # cantidad[0]  -  3 -> 10 - 3 = 7
  productos["cantidad"][indice] = calculo # cantidad[0] = 7

  print(productos)
else:
  print("Producto no encontrado!")
````

````python
# Cargar los datos desde una hoja de cálculo de Google
datos = pd.read_csv("https://docs.google.com/spreadsheets/d/1LKY2GTuQJoFSBwt48J88sO36OpUycLE7rrNlz2ecN8c/export?format=csv")
datos
````

````python
# Ver el tipo de dato que contiene cada columna
datos.info()
````

````python
# Reemplazo de los valores de la columna Altura ',' -> '.'
datos["Altura"] = datos.Altura.str.replace(',', '.')

# Conversión del tipo de dato String -> Float
datos["Altura"] = datos.Altura.astype(float)
datos.info()
````

````python
# Visualizar la estadística descriptiva
datos.describe()
````

````python
# Hacer datos con los datos, crear nuevas columnas a partir de las del DataFrame
datos["Nombre y Apellido"] = datos["Nombre"] + " " + datos["Apellido"]
datos["IMC"] = datos['Peso'] / datos['Altura']**2
datos
````

````python
# Eliminar filas / columas
datos.drop(["Nombre", "Apellido"], axis = 1, inplace = True) # axis = 1 columnas, axis = 0 filas
datos.drop([0], axis = 0, inplace = True)
datos
````

````python
datos = datos[["Nombre y Apellido", "Edad", "Altura", "Peso", "IMC"]]
datos
````

````python
# Condiciones con DataFrames
# datos[datos["Edad"] > 21]
condicion = datos["Edad"] > 21
datos[condicion]
````

````python
# Gráficas con Pandas (Matplotlib)
datos.plot.bar(x = "Nombre y Apellido", y = ["Edad", "IMC"], color = ["r","c"])
plt.title("Título del gráfico")
plt.ylabel("Edades e IMC")
plt.show()
````
