````python
# Determinar si un estudiante se encuentra en nivel Inicio, Proceso o Logrado.
# Inicio < 11
# En Proceso >=11 && <=14
# Logrado >=15

nota1 = 15
nota2 = 11
nota3 = 13

promedio = (nota1 + nota2 + nota3) / 3

if promedio >= 15:
  print("Logrado")
elif promedio >= 11:
    print("En proceso")
else:
    print("Inicio")
````

El dscto se determina de acuerdo al medio de pago:



*   Efectivo -> 10% dscto
*   Tarjeta de débito -> 8% dscto
*   Tarjeta de crédito -> 5% dscto



````python
pago = 100
medioPago = input("Ingresa el medio de pago: ")

if medioPago == "EFECTIVO":
  print(f"Tienes un descuento de {pago*0.1}")
elif medioPago == "TARJETA DE CREDITO":
  print(f"Tienes un descuento de {pago*0.08}")
else:
  print(f"Tienes un descuento de {pago*0.05}")
````

# Listas

````python
lista = [1, 3, 5, "gato"]
lista.append("perro") # inserta un elemento al final de la lista
lista.insert(0, 7) # inserta un elemento en la posicion deseada (posición, elemento)

lista.pop(-1) # elimina un elemento a partir de su índice/posición
lista.remove("gato") #elimina un elemento de la lista

lista.sort() # ordena la lista

lista.index(3) # muestra el índice de un elemento
lista[1]
````

````python
diccionario = {
    "gato": "cat",
    "perro": "dog",
    "caballo": "horse"
}

diccionario["gato"]

diccionario.get("perro") # devolver el valor de una clave
diccionario.keys()
diccionario.values()

diccionario.items()

diccionario["vaca"] = "cow" # insertar en caso de que la clave no exista / modificar en caso de que la clave exista
diccionario["gato"] = "cat2"

diccionario.pop("gato") # elimina un elemento a partir de su clave/llave

diccionario
````

````python
productos = {
    "nombre": ["cuaderno", "libro"],
    "precio": [9, 50],
    "cantidad": [10, 15]
}

nombre = input("Ingresa el producto: ")

if nombre in productos["nombre"]:
  cant_venta = int(input("Ingresa la cantidad vendida: ")) # 3
  indice = productos["nombre"].index(nombre) # nombre.index("cuaderno") = 0
  calculo = productos["cantidad"][indice] - cant_venta # cantidad[0]  -  3 -> 10 - 3 = 7
  productos["cantidad"][indice] = calculo # cantidad[0] = 7

  print(productos)
else:
  print("Producto no encontrado!")
````
